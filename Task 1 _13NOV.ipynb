```python
import pandas as pd

# Load dataset
df = pd.read_csv("netflix_titles.csv")


# Check basic info
print(df.info())
```

```python
# Identify missing values
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.dropna()      # drops all rows that have ANY missing value
```

```python
print(df.isnull().sum())
```

```python
df.head(3)
```

```python
#df['type'] = df['type'].str.lower() only one coulmn will update so used loop

# Standardize text columns

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.lower()
```

```python
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
# Convert to desired format (string representation)
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
df.head(2)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['release_year'] = df['release_year'].astype('Int64')  # capital I allows missing values
```

```python
text_columns = ['show_id', 'type', 'title', 'director', 'cast', 'country',
                'rating', 'duration', 'listed_in', 'description']

for col in text_columns:
    df[col] = df[col].astype(str)
```

```python
print(df.dtypes)
```

```python
df.head(2)
```

```python
df.to_csv("netflix_cleaned_types.csv", index=False)
```
